```python
import skimage as ski
from skimage import color
import matplotlib.pyplot as plt

# fetch the image
path = r"D:\Image_Processing_Section\scikit-image\Photos\dices.png"
dices = ski.io.imread(path)

# make the image grayscale
grayscaled_dice = color.rgb2gray(dices)


ski.io.imshow(grayscaled_dice)
plt.axis("off")
```

```python
from skimage.filters import threshold_otsu
from skimage.measure import find_contours

# Obtain the optimal tresh value
thresh = threshold_otsu(grayscaled_dice)

# apply thresholding
binarized = grayscaled_dice > thresh

# find contours at a constant value of 0.8
contours = find_contours(binarized, 0.8)


# Create a new figure and axis
fig, ax = plt.subplots()

# Plot each contour
for contour in contours:
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

# Show the image with contours
ax.imshow(grayscaled_dice, cmap='hot')

# Turn off axis
ax.axis('off')

# Show the plot
plt.show()
```

#### Count the dots in a dice's image
- Check that most of the contours aren't bigger in size than 50. If you count them, they are the exact number of dots in the image.

```python
# Set 50 as the maximum size of the dots shape
max_dots_shape = 50

# Count dots in contours excluding bigger than dots size
dots_contours = [cnt for cnt in contours if cnt.shape[0] < max_dots_shape]

# Print the dice's number
print(f"Dice's dots number: {len(dots_contours)}")
```
