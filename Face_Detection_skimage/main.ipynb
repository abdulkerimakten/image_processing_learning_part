```python
import skimage as ski
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

#### Creating Detector

```python
# import the Classifier Class
from skimage.feature import Cascade

# load the trained file from the model root
trained_file = ski.data.lbp_frontal_face_cascade_filename()

# initialize the detector
detector = Cascade(trained_file)
```

# -------------------------------------------------------------

```python
# Function to show the face as marked

def show_defected_face(result, detected, title="Face Image"):
    plt.imshow(result)
    img_desc = plt.gca()
    plt.set_cmap("gray")
    plt.title(title)
    plt.axis("off")

    for patch in detected:
        img_desc.add_patch(
            patches.Rectangle((patch["c"], patch["r"]), patch["width"], patch["height"], fill=False, color="r", linewidth=2)
        )
    
    plt.show()
```

# -------------------------------------------------------------

```python
# apply detector on the image
image = ski.io.imread(r"D:\Image_Processing_Section\scikit-image\Photos\lady.jpg")

detected = detector.detect_multi_scale(image, 
                                       scale_factor=1.3,
                                       step_ratio=1,
                                       min_size=(10,10),
                                       max_size=(400,400))
```

```python
print(detected)
```

```python
# Show image with detected face marked
show_defected_face(image, detected)
```
